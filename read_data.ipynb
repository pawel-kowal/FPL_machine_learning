```python
df = pd.read_csv(r"C:\Users\user\fantasy_premier_league\Fantasy-Premier-League\data\2023-24\gws\merged_gw.csv")
df_raw = df
df_history = df_raw[(df_raw.GW >= 29) & (df_raw.GW<=37)]
df = df[df.GW >= 34]
```

```python
for i in range(1, 6):
    for j in range(1,6):
        df_loop = df[df.GW==j+33]
        df_loop = df_loop.merge(
            df_history[df_history['GW'] == df_loop['GW'].iloc[0] - i],
            on='name',
            suffixes=('', f'_{-i}')
        )
        if j == 1 and i == 1: 
            df_finish = df_loop
        else: 
            df_finish = pd.concat([df_finish, df_loop], ignore_index=True)
df_finish.tail(20)
```
